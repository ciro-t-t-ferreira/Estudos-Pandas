## Imóveis Residenciais

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel.csv', sep = ';')
```

```python
type(dados)
```

```python
list(dados['Tipo'].drop_duplicates()) #Lista cada ocorrência distinta no campo 'Tipo'
```

```python
residencial = ['Quitinete', 
'Casa',
'Apartamento',
'Casa de Condomínio',
'Casa de Vila']
```

```python
dados['Tipo'].isin(residencial) #Compara as entradas da coluna tipos da DataFrame dados com os elementos da lista 'residencial', se ela estiver ali retorna True, do contrário retorna False
```

```python
selecao = dados['Tipo'].isin(residencial)
```

```python
dados_residenciais = dados[selecao] #Isso faz com que apenas as entradas True sejam colocadas na DF dados_residenciais
dados_residenciais
```

```python
dados_residenciais.index = range(dados_residenciais.shape[0]) #Acerta os índices
```

# Exportando a Base de Dados

```python
dados_residenciais.to_csv('dados/aluguel_residencial.csv', sep = ';')
```
