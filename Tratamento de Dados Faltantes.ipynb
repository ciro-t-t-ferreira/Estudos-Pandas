## Tratamento de Dados Faltantes

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel_residencial.csv', sep = ';')
```

```python
#Detectando dados nulos no DataFrame:
dados.isnull() #retorna True para todo dado nulo
```

```python
dados.notnull()#retorna False para todo dado nulo
```

```python
dados.info() #Subtraindo o número de entradas do número de contagens de não nulo conseguimos saber quantos valores nulos existem em cada coluna
```

```python
#Queremos um DataFrame apenas com as entradas de valor nulo
dados['Valor'].isnull() #retorna uma tabela com True apenas nos campos nulos de 'Valor'
```

```python
dados[dados['Valor'].isnull()] #usamos a Series acima para selecionar apenas os registros Nulos no valor na tabela
```

```python
#Queremos descobrir o número de registros nulos retirados:
A = dados.shape[0] #Número de registros da tabela original
dados.dropna(subset = ['Valor'], inplace = True) #Função que exclui os nulos, importante setar inplace = True para substituir a tabela
B = dados.shape[0] #Novo número de registros
A - B
```
