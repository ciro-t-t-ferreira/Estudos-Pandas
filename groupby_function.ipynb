```python
import pandas as pd
```

```python
df = pd.read_csv('aluguel.csv', sep = ';')
df.head(10)
```

## Single Group, Single Column

```python
df.groupby(['Tipo'])['Quartos'].mean() 
#No argumento do groupby entra a coluna pela qual você vai agrupar os valores, a segunda são os valores que serão
#mostrados, e ao final precisa dizer de qual forma eles vão ser aglutinados (nesse caso tenho uma média).
#Funções: min(), max(), mean(), meridian(), count()
```

## Single Group, Multiple Column

```python
df.groupby(['Tipo'])['Quartos','Area'].mean() 
```

## Multiple Groups as Columns

```python
df.groupby(['Tipo','Bairro'])['Quartos'].mean() 
```

## Multiple Functions

```python
df.groupby(['Tipo'])['Quartos'].agg(['mean','sum']) 
```

## Tip: como agrupar variáveis contínuas

```python
#A função cut irá segmentar os registros, nos argumentos você colocar a variável segundo a qual será feito o corte, , o número de cortes que serão feitos e uma lista com os nomes de cada categoria
df['Tamanhos'] = pd.cut(df['Area'], bins = 5, labels = ['Minúscula','Pequena','Média','Grande','Enorme'])
df
#Esse dataframe é um péssimo exemplo, porque temos poucas áreas enormes, que fazem todas as outras serem caracterizadas como minúsculas.
```
