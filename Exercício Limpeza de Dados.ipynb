```python
import pandas as pd
```

```python
imoveis = pd.DataFrame([['Apartamento', None, 970, 68], 
                        ['Apartamento', 2000, 878, 112], 
                        ['Casa', 5000, None, 500], 
                        ['Apartamento', None, 1010, 170], 
                        ['Apartamento', 1500, 850, None], 
                        ['Casa', None, None, None], 
                        ['Apartamento', 2000, 878, None], 
                        ['Apartamento', 1550, None, 228], 
                        ['Apartamento', 2500, 880, 195]], 
                        columns = ['Tipo', 'Valor', 'Condominio', 'IPTU'])
imoveis
```

```python
#Elimina os registros que não apresentam a variável Valor:
imoveis1 = imoveis.copy() #Faço isso para sempre manter o DF original
imoveis1.dropna(subset = ['Valor'], inplace = True)
imoveis1
#VERDADEIRO!
```

```python
#Elimina os imóveis do tipo Apartamento que não apresentam valor Condominio:
imoveis2 = imoveis.copy()
selecao = (imoveis2['Tipo'] == 'Apartamento') & (imoveis2['Condominio'].isnull())
imoveis2 = imoveis2[~selecao]
imoveis2
#Originalmente estava errado, tive que adicionar o ~ na seleção
```

```python
#Substitui os valores faltantes que restam nas variáveis Condominio e IPTU por zero:
imoveis3 = imoveis.copy()
imoveis3 = imoveis3.fillna({'Condominio': 0, 'IPTU': 0})
imoveis3
#Verdadeiro
```
